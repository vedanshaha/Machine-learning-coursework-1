```python
# imported all the needed library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('task2.csv')
df.head()
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,2]
x,y
```

```python
x = np.array(x)
y = np.array(y)
```

```python
# scatter plot of two independent variables on x and y axis
plt.scatter(x[:,0], x[:, 1], c=y, cmap='Blues', edgecolors='black')
plt.title('X1 vs X2')
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=22)
```

```python
best_para = {'alpha': [ 0.001, 0.0002, 0.01, 0.005, 0.5 ,0.2, 0.03 ,0.02]}

from sklearn.model_selection import GridSearchCV

mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100,))

clf = GridSearchCV(mlp, best_para)
clf.fit(x_train, y_train)
```

```python
#The best value for alpha is 0.001, we now use this to fit our classifier

print('Best parameters found:\n', clf.best_params_)
```

```python
mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100,), alpha=0.001)
mlp.fit(x_train, y_train)
```

```python
y_pred1 = mlp.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

print(classification_report(y_test, y_pred1))
```

```python
cm = confusion_matrix(y_test, y_pred1)
ConfusionMatrixDisplay(confusion_matrix = cm,display_labels=mlp.classes_).plot()
plt.show()
```

```python
mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100,), alpha=22)
mlp.fit(x_train, y_train)
y_pred2 = mlp.predict(x_test)
print(classification_report(y_test, y_pred2))

cm = confusion_matrix(y_test, y_pred2)
ConfusionMatrixDisplay(confusion_matrix = cm,display_labels=mlp.classes_).plot()
plt.show()
```

```python
plt.scatter(x_test[:,0], x_test[:, 1], c=y_pred1 ,cmap='Blues', edgecolors='black')
plt.title('scatter plot when alpha is  = 0.001'    )
plt.show()
```

```python
#graph for alpha 

plt.scatter(x_test[:,0], x_test[:, 1], c=y_pred2  ,cmap='Blues', edgecolors='black')
plt.title('scatter plot when alpha is = 22')
plt.show()
```

# Conclusion:

1. Alpha is a parameter for regularization term, aka penalty term, that combats overfitting by constraining the size of the weights. Increasing alpha may fix high variance (a sign of overfitting) by encouraging smaller weights, resulting in a decision boundary plot that appears with lesser curvatures. Similarly, decreasing alpha may fix high bias (a sign of underfitting) by encouraging larger weights, potentially resulting in a more complicated decision boundary.

2. The model with higher alpha overfits the dataset and as a result it's accuracy score and f1 score is low. The model with low alpha has high accuracy and is a better classification model.
